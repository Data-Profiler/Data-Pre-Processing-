```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = sns.load_dataset("tips")
df
```

```python
df.sex.unique()
```

```python
df.smoker.unique()
```

```python
fig = plt.figure(figsize = (5, 5))
axes_1 = fig.add_subplot(1,1,1)
axes_1.scatter(x = df['total_bill'], y = df['tip'], s = df['size'] * 10, alpha = 1)
axes_1.set_title("Total Bill Vs Tip Colored by Sex and sized by size")
axes_1.set_xlabel("Total Bill")
axes_1.set_ylabel("Tip")
```

```python
plt.figure(figsize = (5,5))
sns.regplot(x = "tip", y = "total_bill", data = df)
```

```python
plt.figure(figsize = (10,8))
sns.jointplot(x = "total_bill", y = "tip", data = df, kind = "hex",color="gold")
plt.suptitle("Seaborn visualition")
```

```python
plt.figure(figsize = (5,5))

sns.kdeplot(data = df['total_bill'], data2 = df['tip'], shade = True,color="green")
```

```python
plt.figure(figsize = (8,8))
sns.barplot(x = "time", y = "total_bill", data = df)
```

```python
plt.figure(figsize = (10,8))
sns.barplot(x = "day", y = "total_bill", data = df)
```

```python
plt.figure(figsize = (10,8))
sns.boxplot(x = "time", y = "total_bill", data = df)
```

```python
plt.figure(figsize = (10,8))
sns.violinplot(x = "time", y = "total_bill", data = df)
```

```python
plt.figure(figsize = (10,8))
facet = sns.FacetGrid(df, col = "day", hue = "sex")
facet.map(plt.scatter, "total_bill", "tip")

facet.add_legend()
```

```python
plt.figure(figsize = (10,8))
facet = sns.FacetGrid(df, col = "time", hue = "smoker")
facet.map(plt.scatter, "total_bill", "tip")
facet.add_legend()
```

```python
pair_grid = sns.PairGrid(df)
pair_grid = pair_grid.map_upper(sns.regplot)
pair_grid = pair_grid.map_lower(sns.kdeplot)
pair_grid = pair_grid.map_diag(sns.distplot, rug = True)
```

```python
plt.figure(figsize = (10, 8))
ax = sns.distplot(a = df['total_bill'], kde= True, hist = True, rug = True, color = "g")
ax.set_title("Total Bill Histogram with Density plot")
```
